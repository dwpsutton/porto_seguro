```python
import pandas as pd, matplotlib.pyplot as plt, numpy as np
%matplotlib inline
```

```python
X_train= np.load(open('../wp004/train_matrix.bin','rb'))
y_train= np.load(open('../wp004/train_labels.bin','rb'))

X_test= np.load(open('../wp004/test_matrix.bin','rb'))
y_test= np.load(open('../wp004/test_labels.bin','rb'))
```

```python
import lightgbm.sklearn
lgb_params = {}
lgb_params['learning_rate'] = 0.01
lgb_params['n_estimators'] = 1300
#lgb_params['max_depth'] = 10
lgb_params['max_bin'] = 10
lgb_params['subsample'] = 0.8
lgb_params['subsample_freq'] = 10
lgb_params['colsample_bytree'] = 0.8   
lgb_params['min_child_samples'] = 500
lgb_params['num_leaves']= 25
# lgb_params['random_state']= 0
gbm1 = lightgbm.sklearn.LGBMClassifier(**lgb_params)
```

```python
from datetime import datetime
tc= datetime.now()

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

clf= gbm1

folds = list(StratifiedKFold(n_splits=3, shuffle=True, random_state=2016).split(X_train, y_train))

S_train = np.zeros(X_train.shape[0])
for j, (train_idx, test_idx) in enumerate(folds):
    X_trainj = X_train[train_idx]
    y_trainj = y_train[train_idx]
    X_holdoutj = X_train[test_idx]

    print ("Fit %s fold %d" % (str(clf).split('(')[0], j+1))
    clf.fit(X_trainj, y_trainj)
    y_predj = clf.predict_proba(X_holdoutj)[:,1]                

    S_train[test_idx] = y_predj
#     S_test_i[:, j] = clf.predict_proba(T)[:,1]
# S_test[:, i] = S_test_i.mean(axis=1)
```

```python
info= plt.hist(S_train,log=True,bins=30)
```

```python
import sklearn.metrics
prec,recall,thresh= sklearn.metrics.precision_recall_curve(y_train,S_train)
plt.plot(prec,recall)
plt.xlim([0.,0.2])
```
